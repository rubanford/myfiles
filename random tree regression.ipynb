```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('owid-covid-data.csv')
```

```python
df = data[data.location=='India']
```

```python
df1 = df.copy()
```

```python
for i in range(len(df1['tests_units'])):
    print(type(df1['tests_units'].iloc[i]))
```

```python
for i in range(len(df1['tests_units'])):
    df1['tests_units'].iloc[i] = str(df1['tests_units'].iloc[i])
    print(type(df1['tests_units'].iloc[i]))
```

```python
def impute_mean(df):
    for i in df.columns:
        if (df[i].dtype == 'float64') | (df[i].dtype == 'int64'):
            df[i].fillna(df[i].mean(),inplace =True)
def impute_mode(df):
    for i in df.columns:
        if (df[i].dtype == 'object'):
            df[i].fillna(df[i].mode(),inplace =True)
```

```python
impute_mean(df1)
impute_mode(df1)
```

```python
for i in df1.columns:
    print((df1[i].isnull().sum()))
```

```python
import datetime as dt
df1["date"]=pd.to_datetime(df1["date"])
df1["date"]=df1["date"].map(dt.datetime.toordinal)
```

```python
df1
```

```python
rem_cols = []
for i in df1.columns:
#     print(df1[i].dtype)
    if df1[i].dtype == 'object':
        rem_cols.append(i)
df1.drop(rem_cols,axis=1,inplace = True)
```

```python
df1.info()
```

```python
df1.head()
```

```python

X = df1[['date','new_cases','total_tests','new_tests','stringency_index']]
X
```

```python
y = df1.total_cases
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(n_estimators = 35, max_depth=2, random_state=0)
regr.fit(X_train, y_train)
```

```python
from sklearn import datasets, metrics
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
```

```python
regr.score(X_test, y_test)
```

```python
y_pred=regr.predict(X_test)
```

```python
rmse_1 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
rmse_1
```

```python
r2_1 = r2_score(y_test,y_pred)
r2_1
```
