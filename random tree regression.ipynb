```python
def impute_mean(df):
    for i in df.columns:
        if (df[i].dtype == 'float64') | (df[i].dtype == 'int64'):
            df[i].fillna(df[i].mean(),inplace =True)
def impute_mode(df):
    for i in df.columns:
        if (df[i].dtype == 'object'):
            df[i].fillna(df[i].mode(),inplace =True)
```
